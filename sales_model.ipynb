```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data_sales = {
  'month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
  'sales': [2000, 2200, 2300, 2500, 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300]
}
```

```python
df_sales = pd.DataFrame(data_sales)
```

```python
df_sales['month_number'] = range(1, 13)
```

```python
df_sales
```

```python
X = df_sales[['month_number']] # Double brackets to make X a 2D array
y = df_sales['sales'] # Single brackets to make y a 1D array

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print("Coefficient (slope):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
december_input = pd.DataFrame({'month_number': [12]})
december_prediction = model.predict(december_input)
print("Predicted sales for December:", december_prediction[0])
print("Actual sales for December:", df_sales[df_sales['month_number'] == 12]['sales'].values[0])
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df_sales['sales'], bins=6, color='steelblue', edgecolor='black')
plt.title('Sales Histogram')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_sales['month_number'], df_sales['sales'], color='blue', s=100, label='Actual Sales')
```

```python
X_line = pd.DataFrame({'month_number': [1, 12]})
y_line = model.predict(X_line)
plt.plot([1, 12], y_line, color='red', linewidth=2, label='Linear Regression')
```

```python
plt.title('Predicted vs Actual Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(range(1, 13), df_sales['month'], rotation=45)
plt.show()
```
