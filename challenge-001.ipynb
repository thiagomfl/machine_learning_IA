```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
revenue = {
  'data_ref': ['2023-01-01', '2020-02-01', '2021-03-01', '2022-04-01', '2023-05-01','2023-06-01',  '2020-07-01', '2021-08-01', '2022-09-01', '2023-10-01','2022-11-01', '2023-12-01'],
  'value': [400000, 890000, 760000, 430000, 920000, 340000, 800000, 500000, 200000, 900000, 570000, 995000]
}
```

```python
df_revenue = pd.DataFrame.from_dict(revenue)
```

```python
df_revenue['data_ref'] = pd.to_datetime(df_revenue['data_ref'])
```

```python
df_revenue.info()
```

```python
df_revenue['value'].mean()
```

```python
# Ordenar por data e criar gr√°fico de barras vertical
df_revenue.sort_values('data_ref').plot.bar(x='data_ref', y='value')
```

```python
# Ordenar por data e criar gr√°fico de linhas
df_revenue.sort_values('data_ref').plot.line(x='data_ref', y='value')
```
